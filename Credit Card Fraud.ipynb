## Normalizing the data

```python
from sklearn.preprocessing import Normalizer
```

```python
normalized=Normalizer()
print(normalized.fit_transform(data))
```

## Handling the imbalance Data

```python
data['Class'].value_counts()
```

```python
legit=data[data['Class']==0]
fraud=data[data['Class']==1]
```

```python
legit.describe()
```

```python
fraud.describe()
```

```python
data.groupby('Class').mean()
```

```python
## UnderSampling to balance the data
legit_sample=legit.sample(n=473)
## Concatenating two data frames after under sampling
new_data=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

## Splitting the Data into Test and Train

```python
from sklearn.model_selection import train_test_split
```

```python
x=new_data.drop(['Class'],axis=1)
y=new_data[['Class']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
```

## Model using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
fit_rf=RandomForestClassifier(random_state=42)
```

### Hyperparameter Optimization using Grid Search CV

```python
np.random.seed(40)
param_dic={'max_depth':[20,22,24],'bootstrap':[True,False],'max_features':['auto','sqrt','log2',None],'criterion':['gini','entropy']}
            
cv_rf=GridSearchCV(fit_rf,cv=10,param_grid=param_dic,n_jobs=3)

cv_rf.fit(x_train,y_train)
print('Best Parameters using grid search: \n', cv_rf.best_params_)
```

```python
fit_rf.set_params(criterion='gini',bootstrap=False,max_depth=20,max_features='log2')
```

```python
fit_rf.fit(x_train,y_train)
```

```python
predictions_rf = fit_rf.predict(x_test)
predictions_rf
```

## Evaluation

```python
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
```

```python
accuracy_rf = fit_rf.score(x_test,y_test)
accuracy_rf
```

```python
print("Model recall is::", (recall_score(y_test, predictions_rf))*100)
```

```python
print(classification_report(y_test,predictions_rf )) 
```
